```scala
%pyspark
from pyspark.sql.functions import col, upper, when
```

```scala
%pyspark
df = spark.read.csv("s3a://hadoopbacket/files/Titanic.csv", header=True, inferSchema=True)
df.printSchema()
df.show(5)
```

```scala
%pyspark
df_transformed = df \
    .filter(col("Age") > 18) \
    .withColumn("Name_Upper", upper(col("Name"))) \
    .withColumn("Survived_Label", when(col("Survived") == 1, "Выжил").otherwise("Погиб")) \
    .withColumn("Fare_Category", when(col("Fare") > 50, "Дорогой").otherwise("Дешёвый")) \
    .dropDuplicates() \
    .na.drop(subset=["Age", "Embarked"])

df_transformed.show(5)
```

```scala
%pyspark
df_transformed.write \
    .mode("overwrite") \
    .parquet("s3a://hadoopbacket/output/titanic_result/")
```

```scala
%pyspark
df_result = spark.read.parquet("s3a://hadoopbacket/output/titanic_result/")

df_result.printSchema()
df_result.show(10)
```
